```python
# Random Invoice Contents gerator and save in PDF File
import random
import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
import os

# Define save path
save_path = r"C:\Users\Dell\Desktop\Python Practice\source_datafiles\pdfs\generated_tabular_invoice.pdf"

# Sample product list
products = [
    "Laptop", "Mouse", "Keyboard", "Monitor", "Headphones",
    "USB Cable", "Charger", "Smartphone", "Tablet", "Webcam"
]

# Generate random data
data = []
for i in range(8):  # 8 random products
    product = random.choice(products)
    qty = random.randint(1, 10)
    price = round(random.uniform(10, 500), 2)
    total = qty * price
    data.append([product, qty, price, total])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Product", "Quantity", "Unit Price", "Total Price"])

# Add a final "Total" row
grand_total = df["Total Price"].sum()
df.loc[len(df)] = ["", "", "Grand Total", grand_total]

# Create PDF
doc = SimpleDocTemplate(save_path, pagesize=A4)
styles = getSampleStyleSheet()
elements = []

# Title
elements.append(Paragraph("Invoice", styles["Title"]))
elements.append(Spacer(1, 12))

# Convert DataFrame to list for Table
table_data = [df.columns.tolist()] + df.values.tolist()

# Create table
table = Table(table_data, colWidths=[150, 80, 100, 100])

# Styling
style = TableStyle([
    ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
    ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
    ("ALIGN", (1, 1), (-1, -1), "CENTER"),
    ("FONTNAME", (0, 0), (-1, 0), "Helvetica-Bold"),
    ("BOTTOMPADDING", (0, 0), (-1, 0), 10),
    ("BACKGROUND", (0, 1), (-1, -2), colors.beige),
    ("GRID", (0, 0), (-1, -1), 1, colors.black),
    ("BACKGROUND", (0, -1), (-1, -1), colors.lightgrey),  # Total row highlight
    ("FONTNAME", (0, -1), (-1, -1), "Helvetica-Bold"),
])

table.setStyle(style)
elements.append(table)

# Build PDF
doc.build(elements)

print (table)

print(f"✅ Invoice saved at: {save_path}")
```
